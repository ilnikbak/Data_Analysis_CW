```python
### YOUR CODE HERE ###
main_df['start_date'] = pd.to_datetime(main_df.start_date, utc = True)
main_df['finish_date'] = pd.to_datetime(main_df.finish_date, utc = True)
main_df['начало']=pd.to_datetime(main_df['start_date'],  yearfirst=True, errors='coerce')\
                                                                            .dt.to_period("M")#2.1 create a new column
main_df['конец']=pd.to_datetime(main_df['finish_date'],  yearfirst=True, errors='coerce')\
                                                                            .dt.to_period("M")#2.1 create a new column
```

```python
main_for_sum = main_df[(main_df['is_homework'] == True) & (main_df['module_number'] == 1)]
sum_of_month = main_for_sum[['title','начало','student_id']].groupby(['title','начало'])\
                                                                            .nunique().unstack(fill_value=0).stack()#2.1 group values by id & summ
sum_of_month.drop(sum_of_month.columns[[0,1]], axis=1, inplace=True) 
sum_of_month.reset_index(inplace = True)

sum_of_month2 = sum_of_month.pivot(index='начало', columns='title', values='student_id').fillna(0)
sum_of_month2.reset_index(inplace = True)

sum_of_month.head(30)
```

```python
from pandas.tseries.offsets import MonthEnd, MonthBegin, Day

main_for_progress = main_df[(main_df['is_homework'] == True)]
sum_of_progress = main_for_progress.copy()
sum_of_progress['start_date'] = sum_of_progress['start_date'].apply(lambda x: pd.Timestamp(x) + Day() - MonthBegin())

sum_of_progress['finish_date'] = sum_of_progress[['start_date','finish_date'
                                                 ]].apply(lambda x: pd.Timestamp(x['start_date'])
                                        - Day() + MonthEnd() if pd.isna(x['finish_date']) else x['finish_date'],axis=1)

sum_of_progress = sum_of_progress[['start_date','finish_date','title',"is_homework",'progress_id']]\
                                                        .groupby(['title','start_date','finish_date',"is_homework"]).nunique()
sum_of_progress.drop(sum_of_progress.columns[[0,1,2,3]], axis=1, inplace=True) 
sum_of_progress.reset_index(inplace = True)
sum_of_progress = sum_of_progress[sum_of_progress["is_homework"] == True]
```

```python
#Задание 2.4
from math import ceil
sum_of_progress['дельта'] = sum_of_progress[['start_date','finish_date']
                           ].apply(lambda x: list(pd.date_range(start = x['start_date'],
                                                                periods = ceil((x['finish_date'] - x['start_date']
                                                                                           )/np.timedelta64(1, 'M')
                                                                                                       ), freq='M')
                                                 ),axis=1)
lst_col = 'дельта'
for_linegraph = pd.DataFrame({col:np.repeat
                         (sum_of_progress[col].values, sum_of_progress[lst_col].str.len()) for col in sum_of_progress.columns.difference(
                             [lst_col])}).assign(**{lst_col:np.concatenate(
    sum_of_progress[lst_col].values)})[sum_of_progress.columns.tolist()] = pd.DataFrame(
    {col:np.repeat(sum_of_progress[col].values, sum_of_progress[lst_col].str.len()) for col in sum_of_progress.columns.difference(
        [lst_col])}).assign(**{lst_col:np.concatenate(sum_of_progress[lst_col].values)})[sum_of_progress.columns.tolist()]
```

```python
for_linegraph['дельта'] = for_linegraph['дельта'].dt.to_period('M')
for_linegraphgroup = for_linegraph.groupby(['title','дельта']
                           )['progress_id'].count()
for_linegraphgroup= for_linegraphgroup.to_frame().reset_index()
```

```python
### YOUR CODE HERE ###Отобрать первое и последнее дз сразу Вычитать из i+1 i
#Отбираем домашние работы (первую и последнюю если их в модуле несколько) и оставляем только те модули, где есть ДР
main_for_sort = main_df[['title',
                         'module_number',
                         'lesson_number']][(main_df["is_homework"] == True)].sort_values(by='lesson_number')
last_hw = main_for_sort.groupby(['title','module_number'],as_index=False).last()
first_hw = main_for_sort.groupby(['title','module_number'],as_index=False).first()
#Сдвинул на -1 индекс, не очень понял зачем, но без этого не мог соединить адекватно
first_hw= first_hw.shift(periods=-1)
First_and_last_hw = last_hw.merge(first_hw, left_index=True, right_index=True, suffixes=('_lst', '_fst'))
#Убрал первые модули курса, которые подтянулись к первым модулям курса
First_and_last_hw = First_and_last_hw[First_and_last_hw['title_lst'] == First_and_last_hw['title_fst']]
#Дропнул лишний столбик
First_and_last_hw.drop('title_fst', axis='columns', inplace=True)
First_and_last_hw

#Собрал все домашние работы, которые закончены, посчитал количество прогрессов
count_finished_hw = main_df[(main_df['is_homework'] == True) & (main_df['status'] == 'done')].groupby(['title',
                                                                                                       'module_number',
                                                                                                       'module_title',
                                                                                                       'lesson_number'])['progress_id'].count().reset_index()
#Собрал все домашние работы, вне зависимости от статуса, посчитал количество прогрессов у них
count_all_hw = main_df[(main_df['is_homework'] == True)].groupby(['title',
                                                                  'module_number',
                                                                  'module_title',
                                                                  'lesson_number'])['progress_id'].count().reset_index()
# К таблице с отобранными модулями и дом заданиями присоединяем данные по количеству законченных ДЗ
conversion_df = First_and_last_hw.merge(count_finished_hw, 
                                        how='left',
                                        left_on=['title_lst', 'module_number_lst', 'lesson_number_lst'],
                                        right_on=['title', 'module_number', 'lesson_number'])

# К таблице с отобранными модулями и дом заданиями присоединяем данные по количеству начатых, ожидающих ДЗ 
conversion_df = conversion_df.merge (count_all_hw,
                                     how='left',
                                     left_on=['title_lst', 'module_number_fst', 'lesson_number_fst'],
                                     right_on=['title', 'module_number', 'lesson_number'])
# считаем конверсию
conversion_df['conversion'] = conversion_df['progress_id_y'] / conversion_df['progress_id_x']

# Убираем все столбцы кроме нужных
conversion_df = conversion_df[['title_lst', 'module_number_lst', 'module_title_x','conversion']]
conversion_df.column = ['title', 'module_number', 'module_title','conversion']
```
